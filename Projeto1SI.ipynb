#Conjunto de dados
- **AT**(Ambient Temperature): Temperatura em °C. Afeta a densidade do ar e a eficiência da combustão

- **V**(Exhaust Vacuum): Vácuo de escape em cm Hg. Relacionado ` a pressão na turbina a vapor.

- **AP**(Ambient Pressure): Pressão atmosférica em mbar.

- **RH**(Relative Humidity): Umidade relativa em porcentagem.

- **PE**(Net Hourly Electrical Energy Output): A variável alvo em MW (MegaWatts).

#3 - Construa um modelo de regressão linear múltipla utilizando as variáveis independentes (AT, V, AP, RH) para prever a variável dependente (PE). Divida os dados em treino (80%) e teste (20%).

#4 - Avalie o desempenho do modelo utilizando métricas como R², RMSE e MAE. O modelo é adequado para prever a produção de energia? Justifique sua resposta.

```python
import pandas as pd

df = pd.read_excel('UCI CCPP.xlsx')

print(df.head())
```

Como podemos notas temos mais de 2 variaveis, ou seja usaremos a regressão linear múltipla

$$
y = \theta_0 + \theta_1 x_1 + \theta_2 x_2 + ... + \theta_px_p + \epsilon
$$

em que:

$\bullet$ $x_1, x_2, ... , x_p$ são variáveis independentes (Tambem chamdas de características ou features);

$\bullet$ $y$ é a variável dependente (valor no qual desejamos aproximar ou prever);

$\bullet$ $\theta_0$ é o coeficiente linear da reta;

$\bullet$ $\theta_1, \theta_2, ... , \theta_p$ são os coeficientes que representam o impacto de cada variável independente em $y$

$\bullet$ $\epsilon$ é o erro (ou resíduo) da aproximação


Usamos todas as variaveis para esse cálculo por tanto nosso `x` e `y` é dado por:

```python
#Removendo a coluna PE do dataframe x
x = df.drop(columns=['PE'])
#data frame da coluna excluida (y)
y = df['PE']
```

Como podemos notar, temos uma grandeza de tamanhos bem diferente entre nossas variáveis, com isso temos que usar uma normalização `(Padronição z-score, Normalização (Min-Max scaling))`
  
- Padronização z-score = $\frac{x - 𝝁}{\sigma}$
- Normalização (Min-Max scaling) = $x_{norm} = \frac{x - x_{min}}{x_{max}- x_{min}}$

Vamos analisar os 2


```python
#Padronização z - score
from sklearn.preprocessing import StandardScaler
z_score = StandardScaler()
#Ajustar e transformar os dados
x_z_score = z_score.fit_transform(x)  #fit() = calcula: média (μ) e desvio padrão (σ) e transform() aplica a formula

print(x_z_score)
```

```python
#Normalização (Min-Max scaling)
from sklearn.preprocessing import MinMaxScaler
min_max = MinMaxScaler()
#Ajustar e transformar os dados
x_min_max = min_max.fit_transform(x)  #fit() = encontra: valor mínimo (xmin) e valor máximo (xmax) transform() aplcia na formula

print(x_min_max)

print(x_min_max.min())
print(x_min_max.max())
```

Resultados:

z-score = $[-\infty, \infty]$

Min-Max-sacaling = [0, 1]

Melhores ocasiões para o uso de cada um dela

|Critérios|Z-score|Min-Max-Scaling|
|:---|:---|:---|
|Outliers|Lida Melhor|Sensível|
|Distribuição|Normal(Ideal)|Qualquer|
|Interpretação|Média = 0|Escala fixa|
|Uso comum|Estatística/ML|Deep Learning|

O Z-score é mais adequado quando há presença de outliers e quando se deseja padronizar os dados com média zero e desvio padrão um. Já o Min-Max Scaling é utilizado quando se deseja limitar os dados a um intervalo específico, geralmente entre 0 e 1.

Nesse momento vamos dividir as amostras em treino e teste

```python
from sklearn.model_selection import train_test_split

#Dividindo os dados em treino e teste

x_train, x_test, y_train, y_test = train_test_split(
    x, y,
    test_size=0.2, # 20% dos dados são testes e 80% são treinos
    random_state=42 #Garante que a divisão seja sempre igual
)

print(f"Tamanho do datasheet: {len(df)}")
print(f"Tamanho do dataset de treino: {len(x_train)}")
print(f"Tamanho do dataset de teste: {len(x_test)}")
```

Por fim vamos usar o modelo de regressão linear disponivel na `sklearn`

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)

#model.score = qualidade do modelo
print("R2 treino = ",model.score(x_train, y_train))
print("R2 teste = ",model.score(x_test, y_test))
```

interpretação prática do $R^2$
|Situação|Diagnóstico|
|:---|:---|                                      
|Treino alto, teste baixo|Overfitting|
|Treino $\approx$ Teste|Modelo bom|
|Ambas baixa|Modelo ruim|

Satisfação
|R2|>|0,8|Modelo muito bom|
|:---|:---|:---|:---|
|R2|Entre|0,6 e 0,8|Razoavel|
|R2|<|0,5|Fraco|

Determinando os coefs $\theta$

```python
theta = model.coef_
print(theta)
```

$\hat{y} = -1.98589969 - 0.23209358x_1 + 0.06219991x_2 - 0.15811779x_3$

Formula
$$
R^2 = 1 - \frac{SQR}{SQT} → 1 - \frac{\sum^{N_t - 1}_{i=0}(y^{(i)}- \bar{y})}{\sum^{N_t - 1}_{i=0} (y^{(i)}- \hat{y}^{(i)})^2}
$$

- **SQR:** soma dos quadrados dos resíduos: variação que o modelo não conseguio explicar(Diferença entre os valores reais e os valores preditos)
- **SQT:** Soma total dos quadrados, representa a variação total dos dados em relação à média

Erros quadratico médio(MSE)
$$
MSE = 1 - \frac{1}{N_T}\sum^{N_t-1}_{i=0}(y^{(i)}-\hat{y}^{(i)})^2
$$

$$
RMSE = \sqrt{MSE}
$$


```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

y_pred = model.predict(x_test)

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)


print(f"Coeficiente de determinação R2 = {r2:.4f}")
print(f"Erro Quadrático Médio (MSE) = {mse:.4f}")
print(f"Raiz do erro quadratico médio = {rmse:.4f}")
print(f"Erro Absoluto Médio (MAE) = {mae:.4f}")
```

Com esses resultados e pela tabela de satisfaação acima, podemos dizer que é um bom modelo para essa análise.
