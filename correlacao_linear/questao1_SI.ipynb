```python
import pandas as pd
df = pd.read_csv("dataSet.csv")
```

```python
print(df.columns)
```

```python
df.corr()["PE"].drop("PE")
```

```python
X = df[["AT"]]
y = df["PE"]
```

```python
import matplotlib.pyplot as plt

plt.scatter(X,y)
plt.ylabel('Producao de Energia')
plt.xlabel('Temperatura Ambiente')
plt.grid()
plt.show()
```

```python
X = df[["V"]]
y = df["PE"]
```

```python
plt.scatter(X,y)
plt.ylabel('Producao de Energia')
plt.xlabel('Vacuo de Escape')
plt.grid()
plt.show()
```

```python
X = df[["AP"]]
y = df["PE"]
```

```python
plt.scatter(X,y)
plt.ylabel('Producao de Energia')
plt.xlabel('Pressao Atmosferica')
plt.grid()
plt.show()
```

```python
X = df[["RH"]]
y = df["PE"]
```

```python
plt.scatter(X,y)
plt.ylabel('Producao de Energia')
plt.xlabel(' Umidade Relativa')
plt.grid()
plt.show()
```

As variáveis que apresentam maior influência na produção de energia (PE) são AT (temperatura ambiente) e V (vácuo), pois apresentam correlações mais fortes com a variável dependente.

A variável AT se destaca como a mais relevante, com influência negativa significativa, indicando que o aumento da temperatura reduz a produção de energia.

A variável V também possui forte influência, porém apresenta alta correlação com AT, caracterizando multicolinearidade.

As variáveis AP (pressão) e RH (umidade) apresentam influência menor, contribuindo de forma secundária para o modelo.
